```python
def get_statistics(file, quality):
    """
    file: the file storing data
    quality: string, the quality whose statistics we want 
    """
    print('statistics on ' + quality + ':')
    print('max ' + quality + ':')
    maximum = np.max(file)
    print("{:e}".format(maximum))
    print('min ' + quality + ':')
    minimum = np.min(file)
    print("{:e}".format(minimum))
    # decimal_array_3d = np.array([[[Decimal(str(x)) for x in row] for row in matrix] for matrix in file])
    print('percentile on ' + quality+ ' [25%, 50%, 75%]:')
    # percentile of the flattened grid
    percentile1 = np.percentile(file, [25, 50, 75])
    print(percentile1)
    print('percentile on ' + quality+ ' [10%, 90%]:')
    # percentile of the flattened grid
    percentile2 = np.percentile(file, [10, 90])
    print(percentile2)
```
