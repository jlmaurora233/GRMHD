### Plotting Functions

```python
def plot_compare_xy(dumps_Ma, dumps_Sa, title, min1, max1, min2, max2, figx=20, figy=8):
    # up: Ma+0.5; down: Sa+0.5
    # title: str, like 'log_rho'
    # min1, min2: vmin and vmax for MAD models' plots
    # min2, max2: vmin and vmax for SANE models' plots
    fig = plt.figure(figsize=(figx, figy))
    # MAD models
    ax1 = fig.add_subplot(2, 10, 1)
    pplt.plot_xy(ax1, dumps_Ma[0], title, vmin=min1, vmax=max1, window=(-20,20,-20,20))
    ax2 = fig.add_subplot(2, 10, 2)
    pplt.plot_xy(ax2, dumps_Ma[1], title, vmin=min1, vmax=max1, window=(-20,20,-20,20))
    ax3 = fig.add_subplot(2, 10, 3)
    pplt.plot_xy(ax3, dumps_Ma[2], title, vmin=min1, vmax=max1, window=(-20,20,-20,20))
    ax4 = fig.add_subplot(2, 10, 4)
    pplt.plot_xy(ax4, dumps_Ma[3], title, vmin=min1, vmax=max1, window=(-20,20,-20,20))
    ax5 = fig.add_subplot(2, 10, 5)
    pplt.plot_xy(ax5, dumps_Ma[4], title, vmin=min1, vmax=max1, window=(-20,20,-20,20))
    ax6 = fig.add_subplot(2, 10, 6)
    pplt.plot_xy(ax6, dumps_Ma[5], title, vmin=min1, vmax=max1, window=(-20,20,-20,20))
    ax7 = fig.add_subplot(2, 10, 7)
    pplt.plot_xy(ax7, dumps_Ma[6], title, vmin=min1, vmax=max1, window=(-20,20,-20,20))
    ax8 = fig.add_subplot(2, 10, 8)
    pplt.plot_xy(ax8, dumps_Ma[7], title, vmin=min1, vmax=max1, window=(-20,20,-20,20))
    ax9 = fig.add_subplot(2, 10, 9)
    pplt.plot_xy(ax9, dumps_Ma[8], title, vmin=min1, vmax=max1, window=(-20,20,-20,20))
    ax10 = fig.add_subplot(2, 10, 10)
    pplt.plot_xy(ax10, dumps_Ma[9], title, vmin=min1, vmax=max1, window=(-20,20,-20,20))
    # SANE models
    ax11 = fig.add_subplot(2, 10, 11)
    pplt.plot_xy(ax11, dumps_Sa[0], title, vmin=min2, vmax=max2, window=(-20,20,-20,20))
    ax12 = fig.add_subplot(2, 10, 12)
    pplt.plot_xy(ax12, dumps_Sa[1], title, vmin=min2, vmax=max2, window=(-20,20,-20,20))
    ax13 = fig.add_subplot(2, 10, 13)
    pplt.plot_xy(ax13, dumps_Sa[2], title, vmin=min2, vmax=max2, window=(-20,20,-20,20))
    ax14 = fig.add_subplot(2, 10, 14)
    pplt.plot_xy(ax14, dumps_Sa[3], title, vmin=min2, vmax=max2, window=(-20,20,-20,20))
    ax15 = fig.add_subplot(2, 10, 15)
    pplt.plot_xy(ax15, dumps_Sa[4], title, vmin=min2, vmax=max2, window=(-20,20,-20,20))
    ax16 = fig.add_subplot(2, 10, 16)
    pplt.plot_xy(ax16, dumps_Sa[5], title, vmin=min2, vmax=max2, window=(-20,20,-20,20))
    ax17 = fig.add_subplot(2, 10, 17)
    pplt.plot_xy(ax17, dumps_Sa[6], title, vmin=min2, vmax=max2, window=(-20,20,-20,20))
    ax18 = fig.add_subplot(2, 10, 18)
    pplt.plot_xy(ax18, dumps_Sa[7], title, vmin=min2, vmax=max2, window=(-20,20,-20,20))
    ax19 = fig.add_subplot(2, 10, 19)
    pplt.plot_xy(ax19, dumps_Sa[8], title, vmin=min2, vmax=max2, window=(-20,20,-20,20))
    ax20 = fig.add_subplot(2, 10, 20)
    pplt.plot_xy(ax20, dumps_Sa[9], title, vmin=min2, vmax=max2, window=(-20,20,-20,20))
    plt.show()
```
